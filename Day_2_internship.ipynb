# Prediction using mean data

```python
from sklearn.model_selection import train_test_split
X=df_mean.drop('Life expectancy ',axis=1)
y=df_mean['Life expectancy ']
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model1=RandomForestRegressor()
model1.fit(X_train,y_train)
```

```python
y_pred=model1.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Calculate R-squared (R2)
r2 = r2_score(y_test, y_pred)

print(mae,r2)
```

```python
feature_importances = model1.feature_importances_
for feature_name, importance in zip(X_train.columns, feature_importances):
    print(f"{feature_name}: {importance:.{3}}")
```

```python
def insaurance(age,sex,bmi,children,smoker,region,expenses):
    num=10000000/7
    sum=0
    # Age
    if 18<=age<25 :
        sum=sum + num*0.2
    elif 25<=age<40:
        sum=sum + num*0.3
    elif 40<=age<60: 
        sum= sum+num*0.3
    else:
        sum=sum+num*0.2

    # Sex
    if sex=='male' : 
        sum= sum+ num*0.6
    else:
        sum=sum + num*0.4
    # BMI
    if bmi<18:
        sum=sum+num*0.2
    elif 18<=bmi<30:
        sum=sum+num*0.5
    else:
        sum=sum+num*0.3    
    
    # Children
    if children == 0 :
        sum=sum + num*0.1
    elif children == 1:
        sum = sum + num*0.2
    elif children == 2:
        sum = sum + num*0.2
    else:
        sum = sum + num*0.25    

    # Smoker
    if smoker=='yes' : 
        sum= sum+ num*0.3
    else:
        sum=sum + num*0.7

    # Region
    if region == 'southwest' :
        sum=sum + num*0.3
    elif region == 'southeast':
        sum = sum + num*0.2
    elif region == 'northwest':
        sum = sum + num*0.3
    else:
        sum = sum + num*0.2    

    # Expenses
    if expenses<20000 :
        sum=sum + num*0.125
    elif 20000<=expenses<37000:
        sum=sum + num*0.175
    elif 37000<=age<50000: 
        sum= sum+num*0.2
    else:
        sum=sum+num*0.4

    return sum
```
